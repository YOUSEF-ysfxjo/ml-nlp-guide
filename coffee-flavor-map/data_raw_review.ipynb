## 2.1 Normalize and strip

**Task:** From `df['review']`, produce a series/list of strings where each string has: (a) leading/trailing whitespace removed, (b) no empty strings kept. Optionally normalize encoding (e.g. ensure UTF-8) if you saw issues in Task 1.

```python
import string
df_reviews = (df['review']
                      .str.lower()
                      .str.replace(f'[{string.punctuation}]', '', regex=True)
                      .str.strip())
```

## 2.2 (Optional) Split into sentences

**Task:** Decide: use each review as one line (paragraph-level), or split on sentence boundaries (e.g. `. ` or `; `) so each line is one sentence. If you split, flatten the result so you have a list of sentences, not a list of lists.

**What to check:** For Word2Vec, more shorter lines often help. Use a simple rule (e.g. split on `. `) or a library (e.g. nltk.sent_tokenize) if you prefer. Output: a single list (or generator) of text segments.

```python
# Your code: either keep one paragraph per line, or split each review into sentences and flatten to one list of strings
# Your code: Path or os.makedirs for data/processed; open(..., 'w', encoding='utf-8'); write one segment per line
sentences = df['review'].str.split(r'(?<=[.!?])\s+').explode().tolist()

# Clean up any empty strings or purely whitespace entries
sentences = [s.strip() for s in sentences if len(s.strip()) > 0]
sentences
```

## 2.3 Write corpus to disk

**Task:** Write your list of text segments to `data/processed/corpus.txt`, one segment per line. Ensure the `data/processed/` directory exists (create it if needed). Use UTF-8 encoding when writing.

**What to check:** After running, the file exists; opening it shows one review (or one sentence) per line; no extra blank lines unless you intend them.

```python
processed_path = Path(r"C:\Users\HP\Desktop\projects\New_ara\coffee-flavor-map\data\processed")
processed_path.mkdir(parents=True, exist_ok=True)
file_path = processed_path / 'corpus.txt'

with open(file_path, 'w', encoding='utf-8') as f:
    for s in sentences:
        f.write(f"{s}\n")

print(f"Successfully saved {len(sentences)} segments to {file_path}")
```
